### Step 1. Import the necessary libraries

```python
import numpy as np
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called df.

```python
df = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', sep='|')
df.head()
```

### Step 4. Discover what is the mean age per occupation

```python
df.groupby('occupation').agg({'age': 'mean'})
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
def gender_to_numeric(x):
    if x == 'M':
        return 1
    if x == 'F':
        return 0

df['gender_n'] = df['gender'].apply(gender_to_numeric)


a = df.groupby('occupation').gender_n.sum() / df.occupation.value_counts() * 100 

a.sort_values(ascending = False)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
df.groupby('occupation').agg({'age': ['max', 'min']})
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
df.groupby(['occupation', 'gender']).agg({'age': 'mean'})
```

### Step 8.  For each occupation present the percentage of women and men

```python
gender = df.groupby(['occupation', 'gender']).agg({'gender': 'count'})
gender
```

```python
gender_count = df.groupby('occupation').agg('count')
gender_count
```

```python
occup_gender = gender.div(gender_count, level='occupation')
occup_gender
```
